```python
def check_countries(headline):
    countries_in_headline = []
    # Get List of Cities
    gc = geonamescache.GeonamesCache()
    cities = gc.get_countries_by_names()
    # Iterate Over List of Countries
    for country in countries.keys():
        if country in headline:
            countries_in_headline.append(country)
    if not countries_in_headline:
        return np.nan
    return countries_in_headline
```

```python
def check_city(headline):
    cities_in_headline = []
    # Get List of Cities
    gc = geonamescache.GeonamesCache()
    cities = gc.get_cities()
    # Iterate Over List of Cities
    for city_index in cities.keys():
        if cities[city_index]['name'] in headline:
            cities_in_headline.append(cities[city_index]['name'])
    if not cities_in_headline:
        return np.nan
    return cities_in_headline
```

```python
%%time
headline = []
country_hl = []
city_hl = []
file = open("./data/headlines.txt", 'r')
for counter, line in enumerate(file.readlines()):
    if counter % 25 == 0:
        print(f'{counter} completed out of 650')
    # remove 's
    line = line.replace("'s", '')
    # remove punctuation
    line = re.sub(r'[^\w\s]','',line)
    # remove new line
    line = line.replace('\n', '')
    headline.append(line)
    country_hl.append(check_countries(line))
    city_hl.append(check_city(line))
file.close()
```

```python
def clean_city(headline, city):
    cities_test = []
    if (type(city) == list):
        if (len(city) > 1):
            for x in headline.split(' '):
                if x in city:
                    cities_test.append(x)
            # check if any lists have cities with more than 1 word ie Los Angeles
            max_len = max([len(x.split()) for x in city])
            for counter, x in enumerate(headline):
                for j in range(1, max_len+1):
                    my_string = ''
                    for i in range(j):
                        if counter + i < len(headline.split()):
                            my_string = my_string + f' {headline.split()[counter + i]}'
                            my_string = my_string.strip()
                            if my_string in city:
                                cities_test.append(my_string)
        else:
            return city
    res = []
    [res.append(x) for x in cities_test if x not in res]
    return res 
```

```python
new_cities = []
for index, row in df.iterrows():
    new_cities.append(clean_city(df.loc[index, 'Headline'], df.loc[index, 'City']))
```

```python
def final_city(list_cities):
    max_len = 0
    city = ''
    for x in list_cities:
        if len(x) > max_len:
            city = x
            max_len = len(x)
    return city
```

```python
df['new_single_city'] = df['New_Cities'].map(final_city) 
```

```python
def final_country(country):
    if type(country) == float:
        return ''
    else:
        return country[0]
```
